```python
import json
from pathlib import Path

from matplotlib import pyplot as plt

# from graphbrain_semsim.case_studies import get_result_dir
```

```python
# results_dir_path = get_result_dir(subdir='countries_20-most-popul_thresholds')
results_dir_path = Path("../case_studies/results/countries_20-most-popul_thresholds")
```

```python
plot_data = {'threshold': [], 'n_results': []}

for file_name in results_dir_path.rglob('*.json'):
    with open(results_dir_path / file_name) as fp:
        results_dict = json.load(fp)

    plot_data['threshold'].append(results_dict['config']['similarity_threshold'])
    plot_data['n_results'].append(len(results_dict['results']))


    
plt.scatter('threshold', 'n_results', data=plot_data)
plt.xlabel("similarity threshold")
plt.ylabel("number of matches")

plt.savefig('plot_countries_20-most-popul_thresholds.png')
```
